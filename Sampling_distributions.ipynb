```python
#binomial distribution

import numpy as np
import math
from matplotlib import pyplot as plt

arr_mean_bi_5 = np.array([])
arr_var_bi_5 = np.array([])
arr_t_distribution_bi_5 = np.array([])

bins = np.arange(35,65,1)
bins2 = np.arange(10,40,1)
bins3 = np.arange(-10,10,1)

for i in range(50):
    n = 5
    a = np.random.binomial(100, 0.5, size = n)
    mean = np.mean(a)
    var = np.var(a)
    std = np.std(a)
    t_distribution = (mean - 50)/(std*math.sqrt(n))
    
    arr_mean_bi_5 = np.append(arr_mean_bi_5, np.array([mean]))
    arr_var_bi_5 = np.append(arr_var_bi_5, np.array([var]))
    arr_t_distribution_bi_5 = np.append(arr_t_distribution_bi_5, np.array([t_distribution]))
```

```python
plt.hist(arr_mean_bi_5, bins, width = 0.8)
plt.xlabel('Binomial Distribution : Sample Mean, n = 5')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(arr_var_bi_5, bins2, width = 0.8)
plt.xlabel('Binomial Distribution : Sample Variance, n = 5')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(arr_t_distribution_bi_5, bins3, width = 0.8)
plt.xlabel('Binomial Distribution : T_distribution, n = 5')
plt.ylabel('Frequency')
plt.show()
```
